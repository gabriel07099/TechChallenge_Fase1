## <h1 style="color: red;">ANÁLISE PEDIDOS</h1>

## <h2 style="color: white;">Manipulando informações</h2>

### Verificações na base

```python
# Rotula produtos sem nome na categoria
df_products["product_category_name"] = df_products["product_category_name"].fillna("categoria_desconhecida")

# Preenche os valores em branco como "Nao informado"
df_orders_reviews['review_score'] = df_orders_reviews['review_score'].fillna("Nao informado")

# garantir datas
df_orders["order_purchase_timestamp"] = pd.to_datetime(
    df_orders["order_purchase_timestamp"], errors="coerce"
)
```

### Normalizando dados

#### Normalizar base de reviews para ter apenas um review por pedido, considerando o mais recente


```python
df_orders_reviews_normalizado = df_orders_reviews.copy()

# garantir formato datetime
df_orders_reviews_normalizado["review_creation_date"] = pd.to_datetime(
    df_orders_reviews_normalizado["review_creation_date"], errors="coerce"
)

df_orders_reviews_normalizado["review_answer_timestamp"] = pd.to_datetime(
    df_orders_reviews_normalizado["review_answer_timestamp"], errors="coerce"
)

# criar coluna de referência temporal - priorizar data de resposta, se existir, senão data de criação
df_orders_reviews_normalizado["data_review_ref"] = df_orders_reviews_normalizado["review_answer_timestamp"] \
    .fillna(df_orders_reviews_normalizado["review_creation_date"])

# ordenar do mais antigo para o mais recente
df_orders_reviews_normalizado = df_orders_reviews_normalizado.sort_values(
    ["order_id", "data_review_ref"]
)

# manter só o review mais recente de cada pedido
df_orders_reviews_normalizado = df_orders_reviews_normalizado.drop_duplicates(
    subset="order_id",
    keep="last"
).copy()


print("Linhas originais:", len(df_orders_reviews))
print("Linhas normalizadas:", len(df_orders_reviews_normalizado))
print("Pedidos únicos na base normalizada:", df_orders_reviews_normalizado["order_id"].nunique())
```

#### Normalizar base de pagamentos

```python
# Cria uma linha por pedido e soma o valor total pago em cada um; Tambem cria a coluna de meios de pagamento, juntando os tipos distintos em cada pedido, ordenando os tipos alfabeticamente
df_orders_payment_normalizado = (
    df_orders_payment
    .groupby("order_id", as_index=False)
    .agg(
        valor_total_pago=("payment_value", "sum"),
        meios_pagamento=("payment_type", lambda x: ", ".join(sorted(x.dropna().astype(str).unique())))
    )
)
```

## <h4 style="color: purple;">Criando tabelas de cruzamento</h4>

#### Criando tabela de pedido + categoria + produto

```python
# Cruzando base de pedido + categoria
df_produtos_full = df_orders_items.merge(
    df_products[['product_id', 'product_category_name', 'product_photos_qty']],
        on="product_id",
        how="left"
).copy()
```

#### Criando tabela de valor total de frete por pedido

```python
df_frete_por_pedido = (
    df_produtos_full
    .groupby("order_id", as_index=False)
    .agg(
        valor_frete=("freight_value", "sum")
    )
)
```

```python
# Verifica se há shipping limit date diferentes para o mesmo order_id
shipping_por_pedido = (
    df_produtos_full
    .groupby("order_id")["shipping_limit_date"]
    .nunique()
    .reset_index(name="qtd_datas")
)

# filtrar pedidos com mais de uma data
pedidos_multiplas_datas = shipping_por_pedido[
    shipping_por_pedido["qtd_datas"] > 1
]

print("Pedidos com múltiplas datas de shipping_limit_date:", len(pedidos_multiplas_datas))
pedidos_multiplas_datas.head()
```

#### Criando tabela de clientes + Região do Pedido

```python
# base inicial orders + customers
df_orders_customers = df_orders.merge(
    df_customers[["customer_id", "customer_unique_id", "customer_state", "customer_city"]],
    on="customer_id",
    how="left"
).copy()
```

#### Criando tabelona

```python
df_full = (
    df_orders_customers.merge(
        df_orders_reviews_normalizado[['review_id', 'order_id', 'review_score', 'review_comment_title', 'review_comment_message', 'review_answer_timestamp']],
        on="order_id",
        how="left"
    ).merge(
        df_orders_payment_normalizado,
        on="order_id",
        how="left"
    ).copy()
)

# Cria coluna de cliente reincidente
df_full["cliente_reincidente"] = np.where(
    df_full.duplicated(subset="customer_unique_id", keep=False),
    1,  0
)
```

## <h5 style="color: yellow;">Aplicando filtros a tabelona</h5>

```python
cond_concluido = (df_full["order_status"] == "delivered")
cond_periodo_desejado = (df_full["order_purchase_timestamp"] >= "2017-01") & (df_full["order_purchase_timestamp"] < "2018-08")

df_full = df_full[cond_concluido & cond_periodo_desejado].copy()
```

#### % de reincidentes por estado

```python
# nível cliente
clientes_estado = (
    df_full
    .drop_duplicates(subset=["customer_unique_id", "customer_state"])
    .groupby("customer_state", as_index=False)
    .agg(
        total_clientes=("customer_unique_id", "nunique"),
        clientes_reincidentes=("cliente_reincidente", "sum")
    )
)

clientes_estado["pct_clientes_reincidentes"] = (
    clientes_estado["clientes_reincidentes"] /
    clientes_estado["total_clientes"]
) * 100


# nível pedido
pedidos_estado = (
    df_full.groupby("customer_state", as_index=False)
    .agg(
        total_pedidos=("order_id", "nunique"),
        pedidos_clientes_reincidentes=("cliente_reincidente", "sum")
    )
)

# pedidos de clientes não reincidentes
pedidos_estado["pedidos_clientes_nao_reincidentes"] = (
    pedidos_estado["total_pedidos"] -
    pedidos_estado["pedidos_clientes_reincidentes"]
)

# %
pedidos_estado["pct_pedidos_clientes_reincidentes"] = (
    pedidos_estado["pedidos_clientes_reincidentes"] /
    pedidos_estado["total_pedidos"]
) * 100

pedidos_estado["pct_pedidos_clientes_nao_reincidentes"] = (
    pedidos_estado["pedidos_clientes_nao_reincidentes"] /
    pedidos_estado["total_pedidos"]
) * 100


# juntar tudo
df_clientes_estado_pedido = clientes_estado.merge(pedidos_estado, on="customer_state")
df_clientes_estado_pedido.sort_values(by="pct_pedidos_clientes_reincidentes", ascending=False)
```

#### Gráfico

```python
# ordenar
grafico_estado = df_clientes_estado_pedido.sort_values("total_pedidos", ascending=False).copy()

# figura
fig, ax1 = plt.subplots(figsize=(14, 6))

# barras (pedidos)
bars = ax1.bar(
    grafico_estado["customer_state"],
    grafico_estado["total_pedidos"]
)

ax1.set_xlabel("Estado")
ax1.set_ylabel("Total de pedidos")
ax1.tick_params(axis="x", rotation=45)

# formatar eixo y
ax1.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))

# aumentar limite do eixo
ax1.set_ylim(0, grafico_estado["total_pedidos"].max() * 1.15)

# adicionar valores nas barras (sempre preto)
for bar in bars:
    altura = bar.get_height()

    if altura > grafico_estado["total_pedidos"].max() * 0.2:
        # barras grandes → dentro
        ax1.text(
            bar.get_x() + bar.get_width() / 2,
            altura * 0.9,
            f"{int(altura):,}",
            ha="center",
            va="top",
            color="black",
            fontsize=9
        )
    else:
        # barras pequenas → fora
        ax1.text(
            bar.get_x() + bar.get_width() / 2,
            altura,
            f"{int(altura):,}",
            ha="center",
            va="bottom",
            color="black",
            fontsize=9
        )

# linha (% reincidência)
ax2 = ax1.twinx()

ax2.plot(
    grafico_estado["customer_state"],
    grafico_estado["pct_clientes_reincidentes"],
    marker="o",
    linewidth=2
)

ax2.set_ylabel("% clientes reincidentes")
ax2.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.1f}%"))

# rótulos da linha
for i, v in enumerate(grafico_estado["pct_clientes_reincidentes"]):
    ax2.text(
        i,
        v,
        f"{v:.1f}%",
        ha="center",
        va="bottom",
        fontsize=9,
        color="black"
    )

# título
plt.title("Pedidos por estado vs taxa de reincidência")

# legendas
ax1.legend(["Total de pedidos"], loc="upper left")
ax2.legend(["% clientes reincidentes"], loc="upper right")

plt.tight_layout()
plt.show()
```

## <h3 style="color: yellow;">ANALISE 2 - SÃO PAULO</h3>

#### Formatando

#### Avaliação

```python
# Avaliação da pontuação de revisão por cliente reincidente e não reincidente
df_sp.groupby("cliente_reincidente")["review_score"].mean()
```

#### Logística

##### Formatando colunas

```python
colunas_data = [
    "order_purchase_timestamp",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for col in colunas_data:
    df_sp[col] = pd.to_datetime(df_sp[col], errors="coerce")
```

##### Criando colunas

```python
# tempo total de entrega
df_sp["dias_entrega"] = (
    df_sp["order_delivered_customer_date"] - df_sp["order_purchase_timestamp"]
).dt.total_seconds() / 86400

# atraso em relação ao prazo prometido
df_sp["dias_atraso"] = (
    df_sp["order_delivered_customer_date"] - df_sp["order_estimated_delivery_date"]
).dt.total_seconds() / 86400

# flag de atraso
df_sp["pedido_atrasado"] = (df_sp["dias_atraso"] > 0).astype(int)
```

###### Criar faixas de tempo de entrega


```python
bins_entrega = [0, 3, 7, 10, 15, 20, 30, 60, np.inf]
labels_entrega = ["0-3", "4-7", "8-10", "11-15", "16-20", "21-30", "31-60", "60+"]

df_sp["faixa_entrega"] = pd.cut(
    df_sp["dias_entrega"],
    bins=bins_entrega,
    labels=labels_entrega
)
```

###### Criar faixas de atraso


```python
bins_atraso = [-np.inf, -5, -2, 0, 2, 5, 10, 20, np.inf]
labels_atraso = [
    "Muito adiantado",
    "Adiantado",
    "No prazo",
    "1-2 atraso",
    "3-5 atraso",
    "6-10 atraso",
    "11-20 atraso",
    "20+ atraso"
]

df_sp["faixa_atraso"] = pd.cut(
    df_sp["dias_atraso"],
    bins=bins_atraso,
    labels=labels_atraso
)
```

### Correlação: review score → probabilidade de recompra

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

# cópia da base
base_modelo = df_sp_full.copy()

# garantir uma linha por pedido
base_modelo = base_modelo.drop_duplicates(subset="order_id").copy()

# ordenar por cliente e data para identificar a 1ª compra
base_modelo["order_purchase_timestamp"] = pd.to_datetime(
    base_modelo["order_purchase_timestamp"], errors="coerce"
)

base_modelo = base_modelo.sort_values(
    ["customer_unique_id", "order_purchase_timestamp"]
).copy()

base_modelo["ordem_compra"] = (
    base_modelo.groupby("customer_unique_id").cumcount() + 1
)

# manter apenas a 1ª compra de cada cliente
base_modelo = base_modelo[base_modelo["ordem_compra"] == 1].copy()

# criar frete proporcional
base_modelo["frete_pct"] = (
    base_modelo["valor_frete"] / base_modelo["valor_total_pago"]
)

# tratar infinitos e valores inválidos
base_modelo["frete_pct"] = base_modelo["frete_pct"].replace([np.inf, -np.inf], np.nan)

# features do modelo
features = [
    "review_score",
    "dias_entrega",
    "dias_atraso",
    "frete_pct",
    "valor_total_pago"
]

# target: 1 = cliente reincidente, 0 = não reincidente
target = "cliente_reincidente"

df_model = base_modelo[features + [target]].dropna().copy()

X = df_model[features]
y = df_model[target]

# padronização
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# modelo
model = LogisticRegression(class_weight="balanced", random_state=42, max_iter=1000)
model.fit(X_scaled, y)

# coeficientes
importancia = pd.Series(model.coef_[0], index=features).sort_values()

# cores
cores = ["#E24B4A" if v < 0 else "#1D9E75" for v in importancia]

# gráfico
plt.figure(figsize=(9, 4.5))
ax = importancia.plot(kind="barh", color=cores)

plt.title("Fatores associados à reincidência do cliente")
plt.xlabel("Coeficiente padronizado")
plt.ylabel("Variáveis")
plt.axvline(0, color="black", linewidth=0.8)

# rótulos nos valores
for i, v in enumerate(importancia.values):
    ax.text(
        v + (0.002 if v >= 0 else -0.002),
        i,
        f"{v:.3f}",
        va="center",
        ha="left" if v >= 0 else "right"
    )

plt.tight_layout()
plt.show()

"""
A análise indica que o principal fator associado à não reincidência é o atraso na entrega, sendo este o elemento de maior impacto negativo.
Em contrapartida, o tempo total de entrega apresenta associação positiva, sugerindo que o cliente tolera prazos mais longos desde que estes sejam cumpridos.
O valor do pedido demonstra impacto negativo, indicando que compras de maior valor tendem a ser menos recorrentes.
Já a variável de avaliação não se mostrou relevante no modelo, possivelmente devido à sua correlação com fatores operacionais, como atraso.
"""
```

### Análise de corte (a mais impactante visualmente para investidores)


```python
# 1. Data da primeira compra por cliente
primeira_compra = (
    df_sp_full.groupby("customer_unique_id")["order_purchase_timestamp"]
    .min()
    .dt.to_period("M")
    .rename("corte")
)

# 2. Juntar corte ao df_sp_full
df_corte = df_sp_full.merge(primeira_compra, on="customer_unique_id")
df_corte["mes_pedido"] = df_corte["order_purchase_timestamp"].dt.to_period("M")

# 3. Diferença em meses entre o corte e o pedido
df_corte["mes_indice"] = (
    df_corte["mes_pedido"] - df_corte["corte"]
).apply(lambda x: x.n)

# 4. Montar tabela de retenção
tabela = (
    df_corte.groupby(["corte", "mes_indice"])["customer_unique_id"]
    .nunique()
    .unstack()
    .fillna(0)
)

# 5. Normalizar pela corte inicial (mes 0)
tabela_pct = tabela.divide(tabela[0], axis=0) * 100

# 6. Heatmap
import seaborn as sns
plt.figure(figsize=(16, 8))
sns.heatmap(tabela_pct.iloc[:, :12], annot=True, fmt=".0f", cmap="YlOrRd_r",
            linewidths=0.5, cbar_kws={"label": "% retidos"})
plt.title("Taxa de retenção por corte mensal")
plt.xlabel("Meses após primeira compra")
plt.ylabel("Corte (mês da 1ª compra)")
plt.tight_layout()
plt.show()
```
